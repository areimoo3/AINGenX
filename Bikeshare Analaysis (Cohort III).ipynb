```python
data = data.drop(columns = ['Unnamed: 0.2'])
```

```python
data.head()
```

```python
#replace null values in holiday column with "Not a Holiday"
data ['holiday'] = data['holiday'].fillna('Not a Holiday')

data['holiday'].value_counts()
```

```python
#rename duration to duration (in seconds)
data = data.rename (columns = {'Duration': 'Duration (in seconds)'})

data.head()
```

```python
#add a column for duration in mins
data ['Duration (in mins)'] = data['Duration (in seconds']/60

data.head()
```

```python
data['Duration (in mins)'].mean()
```

```python
#threshold


#Q1 = 25th percentile
#Q3 = 75th percentile 
# Interquartile Range = Q3-Q1
#Median = 50th percentile 

IQR = 19.716667 - 6.716667
IQR
```

```python
data['Duration (in mins)'].describe()
```

```python
Q1 = data['Duration (in mins)'].quantile (0.25)
Q3 = data['Duration (in mins)'].quantile (0.75)

IQR = Q3-Q1

lower_bound = Q1  - 1.5*IQR
upper_bound = Q3 + 1.5* IQR

```

```python
#I want data points that are above the lower bound but below the upper bound

filtered_df = data[(data['Duration (in mins)'] >= lower_bound) & (data['Duration (in mins)'] <= upper_bound)]
```
